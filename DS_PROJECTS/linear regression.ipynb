```python
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv(r'C:\Users\DELL\Downloads\headbrain.csv')
print(data.head())
```

```python
X=data["Head Size(cm^3)"]
Y=data["Brain Weight(grams)"]
plt.plot(X,Y)
```

```python
mx=np.mean(X)
my=np.mean(Y)
n=len(X)
num=0
den=0
for i in range(n):
    num += (X[i]-mx)*(Y[i]-my)
    den += (X[i]-mx)**2
m=num/den
c=my-m*mx
print(m,c)
```

```python

x=np.linspace(np.max(X)+100,np.min(X)-100)
y=m*x+c
plt.plot(x,y,label="Regression Line",color="red")
plt.scatter(X,Y,label="Values",color="black")
plt.legend()
plt.xlabel("Head Size")
plt.ylabel("Brain Weight")
plt.show()
```

```python

numer=0
denom=0

for i in range(n):
    yp =c+m*X[i]
    numer+=(yp-Y[i])**2
    denom+=(my-Y[i])**2
r2=1-numer/denom
print(r2)
```
